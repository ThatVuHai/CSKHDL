```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Đọc dữ liệu từ tệp Excel
df = pd.read_excel('1.xlsm', sheet_name='Customer_Master')

# Lấy dữ liệu kinh độ (Longitude) và vĩ độ (Latitude)
longitude = df['Longitude']
latitude = df['Latitude']

# Tạo một bộ chuẩn hóa (scaler)
scaler = StandardScaler()

# Chuẩn hóa dữ liệu kinh độ và vĩ độ
longitude_normalized = scaler.fit_transform(longitude.values.reshape(-1, 1))
latitude_normalized = scaler.fit_transform(latitude.values.reshape(-1, 1))

# Vẽ biểu đồ của dữ liệu đã chuẩn hóa
plt.scatter(longitude_normalized, latitude_normalized, marker='o', color='blue')

# Đặt tên cho trục tung và trục hoành
plt.xlabel('Normalized Longitude')
plt.ylabel('Normalized Latitude')

# Hiển thị biểu đồ
plt.title('Biểu đồ vị trí chuẩn hóa theo kinh độ và vĩ độ')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tệp Excel
df = pd.read_excel('1.xlsm', sheet_name='Customer_Master')

# Lấy dữ liệu kinh độ (Longitude) và vĩ độ (Latitude)
data = df[['Longitude', 'Latitude']]

# Tạo một bộ chuẩn hóa (scaler) và chuẩn hóa dữ liệu
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)

# Thực hiện K-means clustering với số lượng cụm (clusters) là k

# Tính toán Số lượng bình phương khoảng cách cho các giá trị k
Sum_of_squared_distances = []
kmeans = KMeans(n_clusters=4)
kmeans = kmeans.fit(data_normalized)
Sum_of_squared_distances.append(kmeans.inertia_)

# Thêm cột 'Cluster' vào DataFrame ban đầu để biểu thị cụm tương ứng
df['Cluster'] = kmeans.labels_

# Vẽ biểu đồ cụm
plt.scatter(data_normalized[:, 0], data_normalized[:, 1], c=kmeans.labels_, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.5,marker='*')

# Đặt tên cho trục tung và trục hoành
plt.xlabel('Normalized Longitude')
plt.ylabel('Normalized Latitude')

# Hiển thị biểu đồ
plt.title('Biểu đồ K-means clustering')
plt.show()

# In thông tin về các cụm
cluster_info = df.groupby('Cluster').mean()
print(cluster_info)
```

```python
plt.plot(range(1, 15), Sum_of_squared_distances, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of Squared Distances')
plt.show()
```
